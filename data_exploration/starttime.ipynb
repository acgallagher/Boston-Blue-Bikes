```python
import pandas as pd
import numpy as np
import requests
import zipfile
import io

url = "https://s3.amazonaws.com/hubway-data/202301-bluebikes-tripdata.zip"
r = requests.get(url)
z = zipfile.ZipFile(io.BytesIO(r.content))
df = pd.read_csv(z.open("202301-bluebikes-tripdata.csv"))
df.head()
# df["starttime"] = pd.to_datetime(df["starttime"])
```

## Missing Data

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print(f'{col} - {round(pct_missing*100)}%')
```

## Irregular Data

```python
df.boxplot(column="starttime")
```

```python
df[df["starttime"] > 8.64e+5].sort_values("tripduration")
```

```python
drop_index = df[df["tripduration"] > 4.32e+5].index
df_cleaned = df.drop(drop_index, axis=0)
df_cleaned["tripduration"].describe()
```
