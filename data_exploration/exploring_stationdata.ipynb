```python
df = pd.read_csv("https://s3.amazonaws.com/hubway-data/current_bluebikes_stations.csv", header=1)
df.head()
```

```python
print(df.dtypes)
print("\n")
print(df.info())
print("\n")
print(df.describe())
```

```python
def clean_stationdata(df: pd.DataFrame) -> pd.DataFrame:
    """Optimizes the datatypes for each column."""

    # station id
    df["station id"] = df["Number"].astype("object")
    df = df.drop(["Number"], axis=1)

    # station name
    df["station name"] = df["Name"].astype("object").str.lower()
    df = df.drop(["Name"], axis=1)

    # station latitude
    df["station latitude"] = df["Latitude"]
    df = df.drop(["Latitude"], axis=1)
    df["station latitude"] = pd.to_numeric(
        df["station latitude"], downcast="float", errors="coerce"
    )

    # station longitude
    df["station longitude"] = df["Longitude"]
    df = df.drop(["Longitude"], axis=1)
    df["station longitude"] = pd.to_numeric(
        df["station longitude"], downcast="float", errors="coerce"
    )

    # district
    df["district"] = df["District"].astype("object")
    df = df.drop(["District"], axis=1)

    # public
    df["public"] = df["Public"].astype("object")
    df = df.drop(["Public"], axis=1)

    # total docks
    df["total docks"] = df["Total docks"]
    df = df.drop(["Total docks"], axis=1)
    df["total docks"] = pd.to_numeric(
        df["total docks"], downcast="integer", errors="coerce"
    )

    # deployment year
    df["deployment year"] = df["Deployment Year"]
    df = df.drop(["Deployment Year"], axis=1)
    df["deployment year"] = pd.to_numeric(
        df["deployment year"], downcast="integer", errors="coerce"
    )

    return df

cleaned_df = clean_stationdata(df)
cleaned_df.head()
```

```python
df = cleaned_df
print(df.dtypes)
print("\n")
print(df.info())
print("\n")
print(df.describe())
```
